```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x =np.linspace(0,10,15)
print(x)
y=x**2
print(y)
```

```python
plt.plot(x,y)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
plt.subplot(9,9,1)
plt.plot(x,y)
```

```python
fig=plt.figure(figsize=(5,8),dpi=50)
axis1=fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2=fig.add_axes([0.3,0.5,0.3,0.4])
axis2.plot(x,y)
```
