```python
for i in range (1,501):
    response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=5cfc0bb3ade614aa86578674158c1c46&language=en-US&page={}'.format(i))
    temp_df=pd.DataFrame(response.json()['results'])[['id','title','release_date','popularity','vote_average','vote_count']]
    df = pd.concat([df, temp_df], ignore_index=True)
    
```
