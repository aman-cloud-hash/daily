```python
import matplotlib.pyplot as plt

height = [150,160,170,180]
weight = [50,60,70,80]

names = ["Aman", "Ravi", "Rahul", "Neha"]

plt.scatter(height, weight)

# X-axis labels change karna
plt.xticks(height, names)

plt.title("Height VS Weight")
plt.xlabel("Persons")
plt.ylabel("Weight")

plt.show()
```

```python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr"]
sales = [100,150,120,200]
person = ["Aman","Ravi","Rahul","Neha"]

plt.plot(months, sales)

# Person names points ke paas
for i in range(len(person)):
    plt.text(months[i], sales[i], person[i])

plt.title("Month Sales")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.show()
```

```python
labels = ["Jan-Aman", "Feb-Ravi", "Mar-Rahul", "Apr-Neha"]

plt.plot(labels, sales)

plt.title("Month Sales")
plt.xlabel("Month-Person")
plt.ylabel("Sales")

plt.show()
```
