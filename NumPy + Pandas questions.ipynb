🔹 Q5: Average marks & attendance

```python
import pandas as pd
data = pd.read_csv("students.csv")
```

```python
print("Average Marks:",data["Marks"].mean())
print("Average Attendance:",data["Attendance"].mean())
```

🔹 Q6: Marks > 70

```python
high=data[data["Marks"]>70]
print(high)
```

🔹 Q7: Performance Column Add

```python
def performance(m):
    if m>=80:
        return "Excellent"
    elif m>=60:
        return "Good"
    else:
        return "Poor"
data["performance"]=data["Marks"].apply(performance)
print(data)
```

🔹 Q8: Eligible Column

```python
data["Eligible"]=data.apply(lambda row: "Yes" if row["Marks"]>75 and row["Attendance"]>75 else "No",axis=1)
print(data)
```

🔹 Q9: Descending Sort

```python
sorted_data=data.sort_values(by="Marks",ascending=False)
print(sorted_data)
```
