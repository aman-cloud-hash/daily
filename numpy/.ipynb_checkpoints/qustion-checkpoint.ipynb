🔢 1️⃣ Creating Arrays

Q1: 1 se 10 tak array banao

```python
import numpy as np
arr=np.arange(1,11)
print(arr)
```

Explanation:

np.arange(1,11) → 1 se 10 tak numbers

11 include nahi hota

Q2: 5 zeros ka array banao

```python
arr=np.zeros(5)
print(arr)
```

Explanation:

np.zeros(5) → 5 elements sab 0

Default dtype = float

Q3: 3x3 ones matrix banao

```python
arr=np.ones((3,3))
print(arr)
```

Explanation:

(3,3) → 3 rows, 3 columns

Sab values = 1

📐 2️⃣ Shape, dtype, reshape

Q1: Shape & dtype check karo

```python
arr=np.array([1,5,7,9])
print(arr.shape)
print(arr.dtype)
```

Explanation:

.shape → dimensions

.dtype → data type

Q2: 1D array ko 2x3 matrix me convert karo

```python
arr=np.array([1,2,3,4,5,6])
reshaped=arr.reshape(2,3)
print(reshaped)
```

Explanation:

6 elements → 2 rows × 3 columns

reshape total elements change nahi karta

Q3: -1 use karke reshape karo

```python
arr=np.arange(1,7)
reshaped=arr.reshape(3,-1)
print(reshaped)
```

Explanation:

-1 → NumPy automatically column calculate karega

🔍 3️⃣ Indexing & Slicing

Q1: 3rd element print karo

```python
arr=np.array([5,8,9,6,4,7,8,5])
print(arr[2])
```

Explanation: Index 0 se start hota hai.

Q2: First 3 elements print karo

```python
print(arr[:3])
```

Explanation:

:3 → start se index 2 tak

Q3: 2D matrix ka 2nd row print karo

```python
matrix=np.array([[1,2,3],[4,5,6]])
print(matrix[1])
```

Explanation:

[1] → 2nd row

🟢 4️⃣ Boolean Masking

Q1: 5 se bade numbers print karo

```python
arr=np.array([1,5,8,9])
print(arr[arr>5])
```

Explanation:

Condition true wale elements return

Q2: Even numbers filter karo

```python
print(arr[arr%2==0])
```

Explanation:

%2==0 → even check

Q3: Negative numbers remove karo

```python
arr=np.array([-1,2,-3,4])
print(arr[arr>=0])
```

⚡ 5️⃣ Vectorized Operations

Q1: Sab elements me 5 add karo

```python
arr=np.array([1,5,6])
print(arr+5)
```

Explanation:

Har element pe 5 add

Loop ki zarurat nahi

Q2: Square nikaalo

```python
print(arr**2)
```

Q3: Do arrays add karo

```python
a=np.array([1,5,6])
b=np.array([8,9,6])
print(a+b)
```

Explanation:

Same index values add hote hain

🔄 6️⃣ Broadcasting

Q1: Scalar add karo

```python
arr=np.array([1,2,3])
print(arr+10)
```

Explanation:

10 automatically har element pe apply hota hai

Q2: Row vector add to matrix

```python
matrix=np.array([[1,5,9],[8,9,7]])
print(matrix+np.array([1,1,1,]))
```

Explanation:

Row vector har row pe add hota hai

Q3: Column broadcasting

```python
matrix=np.array([[1,2],[3,4]])
col=np.array([[10],[20]])
print(matrix+col)
```

Explanation:

Column har row me broadcast hota hai

➕ 7️⃣ Mathematical Operations

Q1: Sum

```python
arr=np.array([1,2,3])
print(arr.sum())
```

Q2: Square root

```python
print(np.sqrt(arr))
```

Q3: Exponential

```python
print(np.exp(arr))
```

📊 8️⃣ Mean, Std, Variance

Q1: Mean

```python
print(arr.mean())
```

Q2: Standard Deviation

```python
print(arr.std())
```

Q3: Variance

```python
print(arr.var())
```

Explanation:

Variance = spread

Std = square root of variance

🔗 9️⃣ Dot Product & Linear Algebra

Q1: Dot product

```python
a=np.array([2,3])
b=np.array([3,4])
print(np.dot(a,b))
```

Explanation:

(2×3) + (3×4)

Q2: Matrix multiplication

```python
A=np.array([[1,2],[3,4]])
B=np.array([[5,6],[7,8]])
print(np.dot(A,B))
```

Q3: Identity matrix

```python
I=np.eye(3)
print(I)
```

Explanation:

Diagonal = 1

ML me useful
