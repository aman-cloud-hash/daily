📊 Dataset (Create Using NumPy → Convert to Pandas)

```python
import numpy as np
import pandas as pd

np.random.seed(0)
marks = np.random.randint(30, 100, 10)
attendance = np.random.randint(50, 100, 10)

data = pd.DataFrame({
    "Marks": marks,
    "Attendance": attendance
})

print(data)
```

🧠 Q1: NumPy se Average nikaalo aur Pandas se verify karo

```python
# NumPy average
numpy_avg=np.mean(data["Marks"].values)
# Pandas average
pandas_avg=data["Marks"].mean()
print("Numpy Avg:",numpy_avg)
print("Pandas Avg:",pandas_avg)
```

🔎 Line-by-line Explanation

data["Marks"].values → Pandas column ko NumPy array me convert karta hai

np.mean() → NumPy se average nikalta hai

.mean() → Pandas ka built-in function hai

Dono result same aayega (logic same hai)

🧠 Q2: Marks > NumPy average wale students filter karo

```python
avg = np.mean(data["Marks"].values)
filtered=data[data["Marks"]>avg]
print(filtered)
```

🔎 Explanation

np.mean() → average calculate kiya

data["Marks"] > avg → condition banayi

data[condition] → filtering rows

👉 NumPy calculation + Pandas filtering

🧠 Q2.2: Marks > Pands average wale students filter karo

```python
avg=data["Marks"].mean()
filtered=data[data["Marks"]>avg]
print(filtered)
```

🧠 Q3: Total Score column add karo (Marks + Attendance)

```python
data["Score"]=data["Marks"]+data["Attendance"]
print(data)
```

🧠 Q4: NumPy se Standard Deviation nikaalo

```python
std_numpy=np.std(data["Marks"].values)
print("Standard Deviation:",std_numpy)
```

🔎 Explanation

.values → NumPy array banaya

np.std() → data ka spread measure karta hai

High std → marks zyada scattered

🧠 Q5: NumPy array ko reshape karo (2 columns)

```python
reshaped=data["Marks"].values.reshape(5,2)
print(reshaped)
```

🔎 Explanation

.values → NumPy array

.reshape(5,2) → 5 rows, 2 columns matrix

Useful for ML models

🧠 Q6: Correlation between Marks & Attendance

```python
correlation=np.corrcoef(data["Marks"],data["Attendance"])
print(correlation)
```

🔎 Explanation

np.corrcoef() → correlation matrix return karta hai

1 ke paas → strong positive relation

0 ke paas → no relation

🧠 Q7: NumPy Condition + Pandas Column

Attendance < 75 ho to “Low” warna “Good”

```python
data["Attendance"]=np.where(
    data["Attendance"]<75,
    "low",
    "Good"
)
print(data)
```
