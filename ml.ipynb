```python
from sklearn.linear_model import LinearRegression
```

```python
x=[[1],[2],[3],[4],[5]]
y=[20,40,60,80,100]
```

```python
model= LinearRegression()
model.fit(x,y)
prediction= model.predict([[6]])
print(prediction)
```

```python
from sklearn.model_selection import train_test_split

X = [1,2,3,4,5,6,7,8]

train, test = train_test_split(
    X,
    test_size=0.25,
    random_state=42
)

print("Train:", train)
print("Test:", test)
```
