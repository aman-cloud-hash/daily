```python
import numpy as np
```

mean

```python
data=[10,20,30,40,50]
mean_data=np.mean(data)
print(mean_data)
```

median

```python
data=[10,20,30,40,50]
median_data=np.median(data)
print(median_data)
```

mode

```python
from scipy import stats
```

```python
data=[10,20,30,30,40,50]
mode_data=stats.mode(data)
print(mode_data.mode)
```

variance

```python
data=[2,4,6,8]
variance=np.var(data)
print(variance)
```
